```python
def make_clip(trial_data):

    duration = len(trial_data)
    print(duration, "frames")

    n_bodies = sum(["o"+str(i)+".x" in list(trial_data.columns) for i in range(1, 5)])
    
    # while (len(trial_data) + 1) % 60 != 0:
    #    trial_data = trial_data.append(trial_data.iloc[-1], ignore_index=True)
    make_frame = make_frame_curried(trial_data, n_bodies)
    clip = mpy.VideoClip(make_frame, duration=duration / 60)
    return clip, trial_data

def get_replay(replays, include_control_replay, include_not_control_replay, step_size=None):
    lengths = [len(trial) for trial in replays]
    control = np.array([(trial.idControlledObject.isin(["object1", "object2"])).sum() for trial in replays])
    
    print(lengths)
    print(control)
    
    if not include_control_replay and not include_not_control_replay:
        return None
    
    if include_control_replay and not include_not_control_replay:
        if sum(control) == 0:
            return None
        random_trial = np.random.choice(np.where(control > 0)[0])
        
    elif not include_control_replay and include_not_control_replay:
        if sum(control) == len(control):
            return None
        random_trial = np.random.choice(np.where(control == 0)[0])
    
    else:
        random_trial = np.random.choice(len(control))
        
    trial_data = replays[random_trial].iloc[::step_size]

    return make_clip(trial_data)
```

```python
replays = read_dataset("replays.h5")
```

```python
display = None
data = get_replay(replays, include_control_replay=True, include_not_control_replay=False)

if data is not None:
    clip, trial_data = data
    display = clip.ipython_display(fps=60)
    
display
```
