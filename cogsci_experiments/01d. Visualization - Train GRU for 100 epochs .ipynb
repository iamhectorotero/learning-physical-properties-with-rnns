```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
from isaac.visualization import plot_lineplot_with_paper_style
```

```python
data_base_directory = "GRU_singlebranch/"
feature_sets_names = ["xy_vxvy", "xy_rtheta", "rtheta", "xy_vxvy_rtheta"]
```

```python
for cl_name in ["mass", "force"]:
    for feature_set in feature_sets_names:
        full_data_directory = data_base_directory + feature_set + "/"
        stats = pd.read_hdf(full_data_directory + cl_name + "_stats.h5")
        ax = plot_lineplot_with_paper_style(None, x="Epoch", y="Loss", data=stats, label="Mass loss")
        ax.set_title(cl_name + " " + feature_set)
        plt.show()
        ax = plot_lineplot_with_paper_style(None, x="Epoch", y="Val Accuracy", data=stats, label="Validation")
        ax = plot_lineplot_with_paper_style(None, x="Epoch", y="Train Accuracy", data=stats, label="Training", ax=ax)
        ax.set_title(cl_name + " " + feature_set)
        plt.ylabel("Accuracy")
        plt.show()
```
