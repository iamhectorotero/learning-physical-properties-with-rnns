```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from isaac.visualization import plot_lineplot_with_paper_style
```

```python
directory = "adding_noise_to_training/"
stats = pd.read_hdf(directory+"stats.h5")
```

```python
stats
```

```python
stats.noise = "σ="+stats.noise.astype(str)
```

```python
ax = plot_lineplot_with_paper_style(directory+"mass_losses.pdf", 
                                    x="Epoch", y="Mass Loss", hue="noise",
                                    data=stats)

plt.show()

ax = plot_lineplot_with_paper_style(directory+"force_losses.pdf", 
                                    x="Epoch", y="Force Loss", hue="noise",
                                    data=stats)
```

```python
ax = plot_lineplot_with_paper_style(directory+"mass_all_val_acc_plot.pdf", 
                                    x="Epoch", y="Mass Val Accuracy", hue="noise", 
                                    data=stats)

plt.show()

ax = plot_lineplot_with_paper_style(directory+"mass_all_train_acc_plot.pdf", 
                                    x="Epoch", y="Mass Train Accuracy", hue="noise", 
                                    data=stats)
```

```python
ax = plot_lineplot_with_paper_style(directory+"force_all_val_acc_plot.pdf", 
                                    x="Epoch", y="Force Val Accuracy", hue="noise", 
                                    data=stats)

plt.show()

ax = plot_lineplot_with_paper_style(directory+"force_all_train_acc_plot.pdf", 
                                    x="Epoch", y="Force Train Accuracy", hue="noise", 
                                    data=stats)
```
