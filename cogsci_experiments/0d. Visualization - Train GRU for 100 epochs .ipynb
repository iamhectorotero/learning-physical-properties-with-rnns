```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
from isaac.visualization import plot_lineplot_with_paper_style
```

```python
directory = "GRU_singlebranch/"
```

# Mass training plots

```python
question_type = "mass"

stats = pd.read_hdf(directory+question_type+"_stats.h5")

ax = plot_lineplot_with_paper_style(directory+question_type+"_losses.pdf", 
                                    x="Epoch", y="Loss",
                                    markers=True, data=stats)

plt.show()

ax = plot_lineplot_with_paper_style(directory+question_type+"_all_val_acc_plot.pdf", 
                                    x="Epoch", y="Val Accuracy", 
                                    data=stats, label="Val Acc")
ax = plot_lineplot_with_paper_style(directory+question_type+"_all_train_acc_plot.pdf", 
                                    x="Epoch", y="Train Accuracy",
                                    data=stats, ax=ax, label="Train Acc")
plt.ylabel("Accuracy")
plt.legend()
plt.show()
```

```python
accuracies = []
for _, seed_df in stats.groupby("seed"):
    accuracies.append(seed_df["Val Accuracy"].max())

print(np.mean(accuracies), np.std(accuracies))
```

# Force training plots

```python
question_type = "force"

stats = pd.read_hdf(directory+question_type+"_stats.h5")

ax = plot_lineplot_with_paper_style(directory+question_type+"_losses.pdf", 
                                    x="Epoch", y="Loss",
                                    markers=True, data=stats, ci="sd")

plt.show()

ax = plot_lineplot_with_paper_style(directory+question_type+"_all_val_acc_plot.pdf", 
                                    x="Epoch", y="Val Accuracy", 
                                    data=stats, label="Val Acc")
ax = plot_lineplot_with_paper_style(directory+question_type+"_all_train_acc_plot.pdf", 
                                    x="Epoch", y="Train Accuracy",
                                    data=stats, ax=ax, label="Train Acc")
plt.ylabel("Accuracy")
plt.legend()
plt.show()
```

```python
accuracies = []
for _, seed_df in stats.groupby("seed"):
    accuracies.append(seed_df["Val Accuracy"].max())

print(np.mean(accuracies), np.std(accuracies))
```
